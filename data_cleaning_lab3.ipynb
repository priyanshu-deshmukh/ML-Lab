```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.describe()
```

```python
df['sepal width (cm)'].iloc[5:11] = np.nan
df['sepal length (cm)'].iloc[3:9] = np.nan
df['petal length (cm)'].iloc[6:12] = np.nan
df['petal width (cm)'].iloc[14:20] = np.nan
```

```python
df
```

```python
df.describe()
```

```python
for col in df.columns:
    df[col] = df[col].fillna(df[col].mean())
df.describe()
```

```python
df
```

```python
sns.boxplot(data=df)
plt.show()
```
